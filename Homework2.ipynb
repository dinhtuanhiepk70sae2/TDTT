```python
#bài 3
a, b, c = map(float, input('Độ dài 3 cách a, b, c của 1 tam giác bất kỳ ,mỗi kí tự cách nhau bởi dấu cách:').split())
if a<=0 or b<=0 or c<=0 or not(a+b>c and a+c>b and b+c>a):
    print('không phải là độ dài 3 cạnh của 1 tam giác')
else:
    a2=a*a
    b2=b*b
    c2=c*c
    loai_tam_giac=''
    if a2+b2 == c2 or a2 + c2 == b2 or b2 + c2 == a2 :
        if a==b or a==c or b==c:
            loai_tam_giac='tam giác vuông cân'
        else:
            loai_tam_giac='tam giác vuông thường'
    elif a==b or a==c or b==c:
        loai_tam_giac='tam giác cân'
    elif a==b and b==c :
        loai_tam_giac='tam giác đều'
    else:
        loai_tam_giac='tam giác thường'
print('là một', loai_tam_giac)
#tính diện tích và chu vi của tam giác
C_cv=a+b+c
p=C_cv/2
S_dien_tich=(p*(p-a)*(p-b)*(p-c))**0.5
print('chu vi của tam giác là :',C_cv)
print('Diện tích của tam giác là :',S_dien_tich)
```

```python
#bai5
a, b, c = map(float, input('nhập a, b, c (cách nhau bằng dấu cách)').split())
debta=b**2 - 4*a*c
if debta<0:
    print(f'phương trình {a}x^2 + {b}x + {c} = 0 vô nghiệm')
elif debta==0:
    x=-b/(2*a)
    print(f'phương trình {a}x^2 + {b}x + {c} = 0 có nghiệm kép x1=x2={x}')
else:
    x1=(-b + debta**0.5)/(2*a)
    x2=(-b - debta**0.5)/(2*a)
    print(f'phương trình {a}x^2 + {b}x + {c} = 0 có 2 nghiệm phân biệt x1={x1}, x2={x2}')
```

```python
#bai8
a, b, m = map(int, input('nhập a, b, m (cách nhau bằng dấu cách)').split())
c, d, n = map(int, input('nhập c, d, n (cách nhau bằng dấu cách)').split())
Det=a*d-c*b
if Det==0:
    print('hệ phương trình vô nghiệm ')
else:
    x = (m*d - b*n) / Det
    y = (a*n - m*c) / Det
    print(f'hệ phương trình có nghiệm x = {x}, y = {y}')
```

```python
#bai9
t_cong_viec=int(input('nhập vào thời gian hoàn thành công việc (giây):'))
gio=t_cong_viec//3600
phut=(t_cong_viec-gio*3600)//60
giay=t_cong_viec-gio*3600-phut*60
print(f'thời gian hoàn thành công việc là: {gio} giờ {phut} phút {giay} giây')
```
